```python
import pandas as pd
import sys 
from pathlib import Path
```

```python
sys.path.append(str(Path.cwd().parent))
raw_path = Path.cwd().parent / 'data' / 'raw'
out_path = Path.cwd().parent / 'data' / 'out'
out_path.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(raw_path / 'abfall_abc.csv', encoding='latin1', delimiter=';', header=None)
df = df[0].str.split(";", expand=True)
df.columns = ['Name', 'Abfallart', 'Hinweis']
df = df[df['Name']!='Name']
```

```python
df.to_csv(out_path / 'abfall_abc_cleaned.csv', index=False)
```
